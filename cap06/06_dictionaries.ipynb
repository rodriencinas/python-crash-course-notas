```python
# A simple Dictionary --> se ve de la siguiente forma (supongamos un personaje de un videojuego)
alien_0 = {'color': 'green', 'points': 5, 'name': 'samu'}
print(alien_0['color']) # --> de esta manera accedemos a los datos guardados en el diccionario
print(alien_0['points']) # --> haciendo referencia a su KEY
```

```python
# Working with dictionaries
# A DICTIONARY in Python es a collection of KEY - VALUE pairs.
# Each KEY is connected to a VALUE
alien_1 = {'color': 'red'} # --> 'color' is a KEY and its associated value is 'red'
```

```python
# Accessing values in a Dictionary
alien_2 = {'name': 'rodri', 'color': 'blue', 'points': 10, 'power': 1000}
print(alien_2['color'])
print(alien_2['points'])
print(alien_2['name'])

#supongamos que jugador elimina al alien_0, y se lleva los puntos de dicho alien
name_death_alien = alien_0['name']
new_points = alien_0['points']
print(f"You killed {name_death_alien.title()}, you earned {new_points} points!")
```

```python
# Adding New Key-Value pairs
alien_2['x_position'] = 25
alien_2['y_position'] = 4
print(alien_2)
```

```python
# Starting with an EMPTY dictionary --> a veces es necesario un diccionario vacío, para ir llenandolo a gusto
alien_3 = {} #--> empty dictionary
alien_3['name'] = 'jordi'
alien_3['color'] = 'pink'
alien_3['points'] = 3
print(alien_3)
```

```python
# Modifying values in a dictionary --> name dictionary[name_key] = new_value
alien_3['name'] = 'sergi' #--> reemplazamos 'jordi' por 'sergi'
#print(alien_3['name'])

#veamos un ejemplo donde combinemos varias cosas
alien_3['speed'] = 'medium'
alien_3['x_position'] = 5
alien_3['y_position'] = 1

print(f'The original position of the alien {alien_3['name'].title()}: {alien_3['x_position']}')
if alien_3['speed'] == 'slow':
    x_increment = 1
elif alien_3['speed'] == 'medium':
    x_increment = 2
else:
    x_increment = 3

#The new position
print(f'The new position of the alien {alien_3['name'].title()} is {alien_3['x_position'] + x_increment}')
```

```python
# Removing KEY-VALUES pairs
del alien_3['points']
alien_3 #--> ya no veo la clave 'points'
```

```python
# A Dictionary of similar objects
favorite_languages = {
    'rodri' : 'python',
    'samuel' : 'c++',
    'yisus' : 'assembler xd',
    'edward' : 'rust',
    }

language = favorite_languages['rodri'].title()
print(f'Rodri\'s favorite language is {language}') #--> esto es mucho más limpio que hacerlo todo en la llamada print()
```

```python
# Using GET() to access values
# For dictionaries, we can use the GET() method to set a DEFAULT VALUE that will be returned if the requested key doesn´t exist
print(alien_3.get('points', 'No point value assigned')) #--> si no encuentra la clave
print(alien_3.get('points')) #--> None by default
```

```python
# Looping through a Dictionary
# Looping through all KEY-VALUE pairs
user_0 = {
    'username': 'rencinas',
    'first': 'rodrigo',
    'last': 'encinas',
}

for key, value in user_0.items(): #--> we need two variables
    print(f'\nKey: {key}')
    print(f'Value: {value}')

for a, b in user_0.items(): #--> we can choose any names i want for these two variables
    print(f'\nKey: {a}')
    print(f'Value: {b}')
```

```python
# If we have specific examples, the variable names can be more descriptive,
# which improves the program's readability.
for name, language in favorite_languages.items():
    print(f"{name.title()}'s favorite language is {language.upper()}")
```

```python
# Looping through all the KEYS in a dictionary --> dictionary.keys()
for name in favorite_languages.keys():
    print(name.title())

# Looping through the keys is actually the default behavior when looping through a dictionary
for name in favorite_languages:
    print(name.upper())
```

```python
favorite_languages

amigos = ['rodri', 'yisus']
for name in favorite_languages.keys():
    print(f'Hi {name.title()}')
    if name in amigos:
        language = favorite_languages[name].title()
        print(f'\t{name.title()}, I see you love {language}')
```

```python
if 'sergi' not in favorite_languages.keys():
    print('Sergi, please take our poll!') # --> if statement checks if 'sergi' is in the sequence
```

```python
# Looping through a dictionary's KEYS in a PARTICULAR ORDER
for name in sorted(favorite_languages.keys()): #--> sorted(favorite_languages.keys()) lo que hace es ordenar la lista generada con las keys
    print(f'{name.title()}, thank you for taking the poll')
```

```python
favorite_languages['judas'] = 'python'
favorite_languages
```

```python
# Looping through all VALUES in a dictionary
print('The following languages have been mentioned on the pool:')
for language in favorite_languages.values():
    print(language.upper())
```

```python
# as we seen, python is repeated. To see each language chosen without repetition, we can use a SET()
print('The following languages have been mentioned on the pool:')
for language in set(favorite_languages.values()):
    print(language.upper())
```

```python
# A list of dictionaries
# cómo puedo gestionar una flota de alienígenas?
# veamos:
alien_0 = {'color': 'green', 'points': 5}
alien_1 = {'color': 'blue', 'points': 10}
alien_2 = {'color': 'red', 'points': 15}

aliens = [alien_0, alien_1, alien_2]
for alien in aliens:
  print(alien)
```

```python
# un ejemplo más realista sería:
aliens = []

# make 30 aliens verdes
for alien_number in range(30):
    new_alien = {'color': 'green', 'points': 5, 'speed': 'slow'}
    aliens.append(new_alien)
    #print(id(new_alien))

# vemos los primeros 5 aliens
for alien in aliens[:5]:
    print(alien)

# veamos cuántos aliens tenemos creados:
print(f'\nTotal number of aliens: {len(aliens)}')
```

```python
# change the color of the aliens
for alien in aliens[:3]:
    if alien['color'] == 'green':
        alien['color'] = 'yellow'
        alien['speed'] = 'medium'
        alien['points'] = 10

# veamos los cambios
for alien in aliens[:5]:
    print(alien)
```

```python
# si quisieramos añadir más características nuevas haríamos:
for alien in aliens[:5]:
    if alien['color'] == 'green':
        alien['color'] = 'yellow'
        alien['speed'] = 'medium'
        alien['points'] = 10
    elif alien['color'] == 'yellow':
        alien['color'] = 'red'
        alien['speed'] = 'fast'
        alien['points'] = 15

# veamos los cambios
for alien in aliens[:10]:
    print(alien)
```

```python
# A list in a dictionary
# volvamos al ejemplo de la pizza
pizza = {
    'crust': 'thick',
    'toppings': ['mushrooms', 'extra cheese'],
}

print(f'You ordered a {pizza['crust']}-crust pizza '
    'with the following toppings:')

for topping in pizza['toppings']:
    print(f'\t{topping}')
```

```python
# veamos el ejemplo con los lenguajes favoritos
favorite_languages2 = {
    'jen': ['python', 'rust'],
    'sarah': ['c'],
    'edward': ['ruby', 'go'],
    'phil': ['python', 'haskell'],
}

for name, languages in favorite_languages2.items():
    if len(languages) == 1:
        print(f"\n{name.title()}'s favorite language is:")
        for language in languages:
            print(f'\t{language.title()}')
    else:
        print(f"\n{name.title()}'s favorite languages are:")
        for language in languages:
            print(f'\t{language.title()}')
```

```python
# A dictionary in a dictionary
# estos ejemplos son un poco más complejos, pero totalmente entendibles
usuarios = {
    'rencinas': {
        'nombre': 'rodrigo',
        'apellido': 'encinas',
        'locacion': 'bajo la viña'
      },

    'jabazan': {
        'nombre': 'javier',
        'apellido': 'bazan',
        'locacion': 'alto comedero'
    },
}

for username, user_info in usuarios.items():
    print(f'\nUsername: {username}')
    nombre_completo = f'{user_info['nombre']} {user_info['apellido']}'
    locacion = user_info['locacion']

    print(f'\tNombre completo: {nombre_completo.title()}')
    print(f'\tLocación: {locacion.title()}')
```
