```python
#when I want to do the same action with every item in a list, I can use PYTHON´S FOR LOOP
magicians = ['alice','david','carolina']
for magician in magicians:
    print(f"{magician.title()}, that was a great trick!")
```

```python
for magician in magicians:
  print(f"{magician.title()}, that was a great trick!")
  print(f"I can´t wait to see your next trick, {magician.title()}.\n")
```

```python
#cualquier linea de código desupués del foor loop que no tiene identación, se ejecuta sin repetición
for magician in magicians:
  print(f"{magician.title()}, that was a great trick!")
  print(f"I can´t wait to see your next trick, {magician.title()}.\n")

print("Thank you, everyone. That was a great magic show!")
```

```python
#making Numerical Lists
# range() function
for value in range(1,5):
    print(value)
```

```python
# Using range() to make a list of numbers --> list() function
list(range(6)) # range() también funciona así: pasándole un solo argumento empezará en 0
```

```python
# podemos indicar con un tercer argumento, a step size when generating numbers
list(range(1,12,2))
```

```python
squares = []
for value in range(1,11):
  square = value ** 2
  squares.append(square)
print(squares)
```

```python
# code more concisely
squares = []
for value in range(1,11):
  squares.append(value ** 2)
print(squares)
```

```python
#Simple statistics with a list of numbers
print(min(squares))
print(max(squares))
print(sum(squares))
```

```python
#LIST COMPREHESION --> allows me yo generate a list in just one line of code
squares2 = [value ** 2 for value in range(1,11)]
print(squares2)
```

```python
# Copying a List
#primero veamos un comportamiento muy curioso
lista1 = list(range(5))
lista2 = lista1
print(lista2)
print(lista1)
```

```python
lista2.append(5)
print(lista2)
print(lista1) #vemos que se modifica lista1 también...
```

```python
#to copy a list, we can make a slice
lista3 = lista1[:]
lista3.append("soy la lista 3")
print(lista3)
print(lista1) #acá vemos cómo no se modifica lista 1, distinto al caso anterior
```

```python
#un ejmplo más real
mis_comidas_fav = ['pizza','pollo','pescado','arroz']
friend_comidas_fav = mis_comidas_fav[:]

print(f"Mis comidas favoritas son:\n {mis_comidas_fav}")
print(f"\nLas comidas favoritas de mi amigo son:\n {friend_comidas_fav}")

#para probar que son listas separadas, agregaremos una comida distinta a cada una
mis_comidas_fav.append('guiso')
friend_comidas_fav.append('ensalada')

print(f"\nMis comidas favoritas son:\n {mis_comidas_fav}")
print(f"\nLas comidas favoritas de mi amigo son:\n {friend_comidas_fav}")
```

```python
# hacer mis_comidas_fav = friend_comidas_fav le indica a Python que asocie la nueva variable
# a la lista que ya se encuentra asociada a la variable mis_comidas_fav
```

```python
# Tuples --> sometimes ill want to create a list of items that cannot change
# python refers to values that cannot change as IMMUTABLE, and IMMUTABLE LIST is called TUPLE
# defining a TUPLE
dimensions = (200,150)
print(dimensions[0])
print(dimensions[1])
```

```python
# veamos que sucede si quiere cambiar el valor de uno de los elementos de la tupla:
dimensions[0] = 1000

#se puede iterar por una tupla igual que en una lista
```

```python
# si quisieramos modificar los valores de una tupla, debemos modificar la tupla entera:
dimensions = (40,20)
dimensions
```
