```python
# POSITIONAL ARGUMENTS
"""Each argument passed to a function in a certain order
 is associated with each parameter of the function,
 respecting that order."""
def describe_pet(animal_type, pet_name):
    print(f'\nI have a {animal_type.title()}')
    print(f"My {animal_type}'s name is {pet_name.title()}")

describe_pet('dog', 'lucho')
```

```python
# Multiple functions calls
describe_pet('cat', 'london')
describe_pet('dog', 'firulais')
describe_pet('turtle', 'gambito')
```

```python
# Order matters in positional arguments
# Podemos tener resultados inesperados si cambiamos el orden en el que pasamos los argumentos
# distinto al que fueron definidos los parámetros
def describe_pet(animal_type, pet_name):
    print(f'\nI have a {animal_type.title()}.')
    print(f"My {animal_type}'s name is {pet_name.title()}")

describe_pet('dog', 'firulais') # correct call
describe_pet('lucho', 'dog') # incorrect call

# Always check to make sure the order of the arguments in my
# function call matches the order of the parameters in the
# functions's definition
```

```python
# Keyword arguments
# Is a NAME-VALUE PAIR THAT PASS TO A FUNCION
describe_pet(animal_type='dog', pet_name='manolo')
describe_pet(pet_name='manolo', animal_type='dog')
#vemos que ambas formas funcionan correctamente
```

```python
# Default values
# When writing functions, we can define a DEFAULT VALUE for each parameter
"""If an argument for a parameter is provided in the function call,
 Python uses the argument value. If not, it uses the parameter's
 default value"""
def describe_pet(pet_name, animal_type='dog'): #--> definimos un default value
    print(f'\nI have a {animal_type.title()}')
    print(f'My {animal_type.title()} name is {pet_name.title()}')

describe_pet('lucho')
describe_pet(pet_name='paco')
describe_pet('paquito', 'lion')
```

```python
# Return values --> "return" statement
# A function doesn't always have to display its output directly
# It can process some data and then return a value or set of values
# RETURN VALUE
def get_formatted_name(first_name, last_name):
    full_name = f'{first_name} {last_name}'
    return full_name.title() #indicamos que la función retornará un valor

form_name = get_formatted_name('rodrigo', 'encinas') #debemos proporcionar una variable para guardar el valor retornado
print(form_name)
```

```python
# Making an argument optional
def get_formatted_name2(first_name, middle_name, last_name):
    full_name = f'{first_name} {middle_name} {last_name}'
    return full_name.title()

form_name2 = get_formatted_name2('rodrigo', 'nicolas', 'encinas')
print(form_name2)
# form_name3 = get_formatted_name2('rodrigo', 'encinas') --> esto generará un TypeError
# Por lo que para convertir middle_name optional, debemos darle an empty default value
```

```python
# código anterior modificado con un argumento opcional
def get_formatted_name2(first_name, last_name, middle_name=''):
    if middle_name:
        full_name = f'{first_name} {middle_name} {last_name}'
        return full_name.title()
    else:
        full_name = f'{first_name} {last_name}'
        return full_name.title()

form_name2 = get_formatted_name2('rodrigo', 'encinas')
print(form_name2)
form_name3 = get_formatted_name2('rodrigo', 'encinas', 'nicolas')
print(form_name3)
```

```python
# Returning a Dictionary
# a function can return any kind of value
def build_person(first_name, last_name):
    person = {'first': first_name, 'last': last_name}
    return person

person1 = build_person('rodrigo', 'encinas')
print(person1)
type(person1)
print(f'The first person created is called {person1['first'].title()} {person1['last'].title()}')
```

```python
# extend the previous function
def build_person(first_name, last_name, age=None):
    person = {'first': first_name, 'last': last_name}
    if age:
        person['age'] = age
    return person

person2 = build_person('caterina', 'mur', 20)
print(person2)
person3 = build_person('maximo', 'encinas')
print(person3)
```

```python
# Using a function with a while loop
def get_formatted_name(first_name, last_name):
    full_name = f'{first_name.title()} {last_name.title()}'
    return full_name

while True:
    print('\nPlease tell me your name: ')
    print('(enter "q" at any time to quit)')

    f_name = input('First name: ')
    if f_name == 'q':
        break
    l_name = input('Last name: ')
    if l_name == 'q':
        break

    formatted_name = get_formatted_name(f_name, l_name)
    print(f'\nHello, {formatted_name}')
```

```python
# Passing a list
def greet_users(names):
    for name in names:
        msg = f'Hello {name.title()}!'
        print(msg)

usernames = ['rodrigo', 'caterina', 'maximo']
greet_users(usernames)
```

```python
# Modifying a list in a function
# ejemplo de fabrica de diseños 3d enviados por usuarios, cadena de montaje...
unprinted_designs = ['phone case', 'robot pendant', 'dodecahedron'] # some designs that need to be printed
complet_models = []

while unprinted_designs:
    current_design = unprinted_designs.pop()
    print(f'Printing model: {current_design}')
    complet_models.append(current_design)

print('\nThe following models have been printed: ')
for complet_model in complet_models:
    print(f'\t{complet_model}')
```

```python
# we can REORGANIZE this code by writing two functions, each of which does one specific job
def print_models(unprinted_designs, completed_models):
    """Simulate printing each design, until none are left
    Move each design to completed_models after printing"""
    while unprinted_designs:
        current_design = unprinted_designs.pop()
        print(f'Printing model: {current_design}')
        completed_models.append(current_design)

def show_completed_models(completed_models):
    print('\nThe following models have been printed: ')
    for completed_model in completed_models:
        print(f'\t{completed_model}')

unprinted_designs = ['phone case', 'robot pendant', 'dodecahedron']
completed_models = []

print_models(unprinted_designs, completed_models)
show_completed_models(completed_models)
```

```python
# Preventing a function from modifying a List ---> pasamos una copia de la lista original
# function_name(list_name[:]) --> a la función pasamos una copia con [:] de la lista original
print_models(unprinted_designs[:], completed_models)
```

```python
# Passing an arbitrary number of arguments
# sometimes i wont know ahead of time how many arguments a function needs to accept
def make_pizza(*toppings): #--> the "#" tells python to make a tuple called toppings, containing all the values this function receives
    """Print the list of toppings that have been requested."""
    print(toppings)

make_pizza('pepperoni')
make_pizza('mushrooms', 'green peppers', 'extra cheese')
```

```python
# modificando un poco el código anterior:
def make_pizza(*toppings):
    """Summarize the pizza we are about to make"""
    print('\nMaking a pizza with the following toppings: ')
    for topping in toppings:
        print(f'\t- {topping}')

make_pizza('pepperoni')
make_pizza('mushrooms', 'green peppers', 'extra cheese')
```

```python
# Mixing Positional and Arbitrary Arguments
def make_pizza2(size, *toppings):
    """Summarize the pizza we are about to make."""
    print(f'\nMaking a {size}-inch pizza with the following toppings: ')
    for topping in toppings:
        print(f'\t-{topping}')

make_pizza2(13, 'peperoni')
make_pizza2(16, 'peperoni', 'mushrooms', 'extra cheese')
```

```python
# Using Arbitrary keyword arguments
def build_profile(first, last, **user_info):
    """Build a dictionary containing everything we know about a user."""
    user_info['first_name'] = first
    user_info['last_name'] = last
    return user_info

user_profile = build_profile('rodrigo', 'encinas',
                             location='buenos aires',
                             profesion='estudiante')
print(user_profile)
```
