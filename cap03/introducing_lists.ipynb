```python
#removing an item using the .pop() method
#the .pop() method removes the last item in a list, but it lets you work with that item after removing it
popped_item = listita.pop()
popped_item
```

```python
popped_item2 = listita.pop()
popped_item2
```

```python
#we can use .pop() method to remove an item from any position in a list by including the index of the item I want to remove
motorcycles = ["yamaha", "suzuki", "bmw"]
print(f"The first motorcycle I owned was a {motorcycles.pop(0).upper()} and the last was {motorcycles.pop().upper()}")
```

```python
#removing an item by value --> if i only know the value of the item I want to remove, I can use the .remove() method
personal_isj.append("nely")
personal_isj.append("adriana")
print(personal_isj)
personal_isj.remove("pedro")
print(personal_isj)
```

```python
# f you could invite anyone, living or deceased, to dinner, who
# would you invite? Make a list that includes at least three people you’d like to
# invite to dinner. Then use your list to print a message to each person, inviting
# them to dinner
invitaciones_cena = ["ricardo", "roxana", "julieta"]
print(f"Invito a cenar a mi amada {invitaciones_cena[1].title()}")
print(f"Invito a cenar a mi amada {invitaciones_cena[2].title()}")
print(f"Invito a cenar a mi amado {invitaciones_cena[0].title()}")

invitado_no_asistira = invitaciones_cena.pop().title()
print(f"{invitado_no_asistira} no podrá asistir a la cena")

```

```python
#sorting a List Permanently with the .sort() method
cars = ["bmw", "audi", "toyota", "subaru"]
cars.sort()
print(cars)
```

```python
#sorting List in reverse-alphabetical order
cars.sort(reverse=True)
print(cars)
```

```python
#sorting a List Temporarily with the .sorted() Function
cars = ["bmw", "audi", "toyota", "subaru"]
print(f"Original order of a cars list is:\n{cars}")
print(f"\nSorted order of a cars list is:\n{sorted(cars)}")
print(f"\nOriginal order of a cars list is:\n{cars}")
print(f"\nSorted reverse order of a cars list is:\n{sorted(cars, reverse=True)}")
print(f"\nOriginal order of a cars list is:\n{cars}")
```

```python
#printing a list in reverse order --> reverse original order of a list
cars.reverse()
print(cars)
```

```python
#finding a length of a list
print(len(cars))
```

RESUMEN
Métodos de Lista:

**.append()**: Añade un elemento al final de la lista.

**.pop():** Elimina y devuelve un elemento de la lista

**.remove()**: Elimina la primera ocurrencia de un valor específico de la lista.

**.insert()**: Inserta un elemento en una posición específica de la lista.

Métodos de Cadena (String):

**.title()**: Convierte la primera letra de cada palabra de una cadena a mayúscula.

**.upper()**: Convierte todos los caracteres de una cadena a mayúsculas.

Funciones Integradas:

**print()**: Muestra la salida en la consola.

**range()**: Genera una secuencia de números.

Sentencias y Operadores:

**=** (Asignación): Asigna un valor a una variable o modifica un elemento de la lista.

**del** (Sentencia del): Elimina un elemento de una lista por su índice.

**f-string** (Cadenas formateadas): Permite incrustar expresiones dentro de cadenas de texto.

* ** * ** (Operador de desempaquetado): Desempaqueta un iterable en argumentos (por ejemplo, *range(10) para crear una lista a partir de un rango).

**[]** (Creación de listas): Se utiliza para crear nuevas listas.

**[]** (Indexación de listas): Accede a elementos individuales de una lista utilizando su índice.

**[:]** (Slicing de listas): Extrae una porción (o 'slice') de una lista.

**.reverse()** --> invierte la lista, no la modifica definitivamente

**.sort()** --> ordena y modifica la lista original, reverse=True ordena al revés
