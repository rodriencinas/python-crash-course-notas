```python
# Letting the user choose when to quit
prompt1 = '\nTell me something, and I will repeat it back to you: '
prompt1 += '\nEnter "quit" to end the program. '

message = '' #--> nos aseguramos de que el chequeo en la condición WHILE sea TRUE
while message != 'quit':
    message = input(prompt1)
    print(message)
```

```python
# código anterior mejorado, para evitar que se muestre la palabra clave de salida
prompt1 = '\nTell me something, and I will repeat it back to you: '
prompt1 += '\nEnter "quit" to end the program. '

message = '' #--> nos aseguramos de que el chequeo en la condición WHILE sea TRUE
while message != 'quit':
    message = input(prompt1)
    if message != 'quit': #--> quick check before displaying the message
        print(message)
```

```python
# Using a FLAG
# for a program that should run only as long as many conditions are true, you can define one variable that
# determinates whether or not the entire program is active --> this variable called a FLAG
active = True #--> flag
while active:
    message = input(prompt1)
    if message == 'quit':
        active = False
    else:
        print(message)
```

```python
# Using BREAK to EXIT a loop
prompt2 = '\nPlease enter the name of a city you have visited: '
prompt2 += '\nEnter "quit" when you are finished. '

while True:
    city = input(prompt2)

    if city == 'quit':
        break
    else:
        print(f'I would love to go to {city.title()}!')
```

```python
# Using CONTINUE in a loop --> we can use CONTINUE statement to return to the beginning of the loop
valor_actual = 0
while valor_actual < 10:
    valor_actual += 1
    if valor_actual % 2 == 0:
        continue

    print(valor_actual)
```

```python
# Using a while loop with lists and dictionaries
# moving items from one list to another, lets look at an example
# start with users that need to be verified
unconfirmed_users = ['alice', 'brian', 'candace']
# empty list to hold confirmed users
confirmed_users = []

# we verify, and we move
while unconfirmed_users:
    current_user = unconfirmed_users.pop()

    print(f'Verifying user: {current_user.title()}')
    confirmed_users.append(current_user)

# once a while loop has finished, we display all verified users
print(f'\nThe following users have been confirmed: ')
for user in confirmed_users:
    print(user.title())
```

```python
# Removing all instances of specific values from a list
listita = [1,2,3,4,5,6,6,7]
listita.remove(2)
print(listita)
```

```python
# to remove all instances of 'cat' value, we can use a while loop
pets = ['dog', 'cat', 'dog', 'goldfish', 'cat', 'rabbit', 'cat']

while 'cat' in pets:
    pets.remove('cat')

print(pets)
```

```python
# Filling a dictionary with user input
responses = {}

# set a flag to indicate that polling is active
polling_active = True

while polling_active:
    name = input('\nWhat is your name? ') #prompt for the person's name and response
    response = input('Which is your favorite color? ')

    responses[name] = response #store the response in the dictionary

    repeat = input('Would you like to let another person respond? (yes/no) ').lower()
    if repeat == 'no':
        polling_active = False
print('\n---Poll Results---')
for name, response in responses.items():
    print(f"{name.title()}'s favorite color is: {response.upper()}")
```
